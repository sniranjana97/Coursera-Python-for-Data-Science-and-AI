Plot a histogram of the column `Age`.


```python
# your code goes here

ax = df.hist(column='Age', bins=25, grid=False, figsize=(12,8), color='#86bf91', zorder=2, rwidth=0.9)

ax = ax[0]
for x in ax:

    # Despine
    x.spines['right'].set_visible(False)
    x.spines['top'].set_visible(False)
    x.spines['left'].set_visible(False)

    # Switch off ticks
    x.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")

    # Draw horizontal axis lines
    vals = x.get_yticks()
    for tick in vals:
        x.axhline(y=tick, linestyle='dashed', alpha=0.4, color='#eeeeee', zorder=1)

    # Remove title
    x.set_title("")

    # Set x-axis label
    x.set_xlabel("Age", labelpad=20, weight='bold', size=12)

    # Set y-axis label
    x.set_ylabel("Value", labelpad=20, weight='bold', size=12)
```

```python

count, bin_edges = np.histogram(df['Age'].dropna().values, 10)

df['Age'].plot(kind='hist', figsize=(8, 5), xticks=bin_edges, bins=10, alpha=0.55, color = 'coral', edgecolor = 'black')

plt.title('Age histogram') # add a title to the histogram
plt.ylabel('Years') # add y-label
plt.xlabel('Age') # add x-label
plt.show()
```

## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
df['Age'].plot(kind='box', figsize=(8, 6))

plt.title('Box plot')

plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

Find out the upper and lower bounds.


```python
# your code goes here
df['ConvertedComp'].describe()
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
q_low = df['ConvertedComp'].quantile(0.01)
q_hi  = df['ConvertedComp'].quantile(0.99)

df_filtered = df[(df['ConvertedComp'] < q_hi) & (df['ConvertedComp'] > q_low)]
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_filtered['ConvertedComp'].median()
```
